```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex
```

```python
def deCasteljau(t , x):
    for i in range (1,len(x)):
        for j in range (0,len(x)-i):
            (a,b) = x[j]
            (c,d) = x[j+1]
            e = (1-t)*a+t*c
            f = (1-t)*b+t*d
            x[j] = (e,f)
    return x[0]
```

```python
t = np.linspace(0, 1, 100)
x = [(0,1),(-1.5,2),(-1.5,-2),(0,-1)]
print("input: ", x)
(a,b) = deCasteljau(t,x)
print("Output plot:")
x = [(0,1),(-1.5,2),(-1.5,-2),(0,-1)]
(c,d) = zip(*x)
plt.scatter(c,d)
plt.plot(c,d, label="Control Polygon")
plt.plot(a,b, label="Bézier Curve, b(t)")
plt.legend()

plt.show()
```
